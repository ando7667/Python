```python
from sympy import *
from sympy.plotting import plot
x = Symbol("x")
f = -18 * x ** 3 + 5 * x ** 2 + 10 * x - 30
f
```

```python
plot(f,(x,-1,1))
```

```python
solveset(f)
```

```python
solveset(f<0)
```

```python
solveset(f>0)
```

```python
result=[-oo, oo]
result[1:1] = solve(diff(f),x)
result
```

```python
res_inc = []
res_dec = []
for i in range(1, len(result)):
    if is_increasing(f, Interval.open(result[i-1],result[i])):
        res_inc.append(f"{result[i-1]}, {result[i]}")
    else:
        res_dec.append(f"{result[i-1]}, {result[i]}")
print(f"убывает {res_dec}")
print(f"возрастает {res_inc}")
```

```python
res_func = solve(diff(f),x)
for i in res_func:
    check = f.subs(x,i)
    if check < 0:
        print(f"локальный минимум {check} от х={i}")
    elif check > 0:
        print(f"локальный макисмум {check} от х={i}")
    elif check == 0:
        print(f"{check} пересечение оси х")
```
